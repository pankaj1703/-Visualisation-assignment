```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


np.random.seed(30)
data = {
    'X': np.random.uniform(-10, 10, 300),
    'Y': np.random.uniform(-10, 10, 300),
    'Z': np.random.uniform(-10, 10, 300)
}

df = pd.DataFrame(data)


fig = plt.figure(figsize=(10, 8))
ax = fig.add_subplot(111, projection='3d')


ax.scatter(df['X'], df['Y'], df['Z'], c='b', marker='o', alpha=0.6)

ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')

ax.set_title('3D Scatter Plot of Data')


plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


np.random.seed(15)
data = {
    'Grade': np.random.choice(['A', 'B', 'C', 'D', 'F'], 200),
    'Score': np.random.randint(50, 100, 200)
}


df = pd.DataFrame(data)

plt.figure(figsize=(10, 6))
sns.violinplot(x='Grade', y='Score', data=df, palette='muted')

plt.title('Distribution of Scores Across Different Grades')
plt.xlabel('Grade')
plt.ylabel('Score')

plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)

sales_pivot = df.pivot_table(index='Day', columns='Month', values='Sales', aggfunc='mean')

plt.figure(figsize=(12, 8))


sns.heatmap(sales_pivot, cmap='YlGnBu', annot=True, fmt=".1f", linewidths=.5)

plt.title('Sales Heatmap Across Different Months and Days')
plt.xlabel('Month')
plt.ylabel('Day')

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y**2))


data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}


df = pd.DataFrame(data)


fig = plt.figure(figsize=(12, 8))
ax = fig.add_subplot(111, projection='3d')

surf = ax.plot_surface(x, y, z, cmap='viridis', edgecolor='none')


ax.set_xlabel('X-axis')
ax.set_ylabel('Y-axis')
ax.set_zlabel('Z-axis')


ax.set_title('3D Surface Plot of z = sin(sqrt(x^2 + y^2))')


fig.colorbar(surf)


plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


np.random.seed(25)
data = {
    'Country': ['USA', 'Canada', 'UK', 'Germany', 'France'],
    'Population': np.random.randint(100, 1000, 5),
    'GDP': np.random.randint(500, 2000, 5)
}


df = pd.DataFrame(data)


plt.figure(figsize=(10, 6))


plt.scatter(
    df['GDP'], df['Population'], 
    s=df['Population']*10,  
    alpha=0.6,  
    c=df['Population'],  
    cmap='viridis',  
    edgecolors='w',  
)

# Add labels for each country
for i, row in df.iterrows():
    plt.text(row['GDP'], row['Population'], row['Country'], fontsize=12, ha='center')

# Set axis labels and title
plt.xlabel('GDP (Billion USD)')
plt.ylabel('Population (Millions)')
plt.title('Bubble Chart: Population vs GDP by Country')

# Show the plot
plt.show()
```
