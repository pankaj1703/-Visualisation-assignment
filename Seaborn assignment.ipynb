```python
import matplotlib.pyplot as plt
import numpy as np


np.random.seed(42)  
x = np.random.rand(100) * 10 
y = 2 * x + np.random.randn(100) * 2 

plt.scatter(x, y, color='blue', label='Data points')


plt.xlabel('X Variable')
plt.ylabel('Y Variable')
plt.title('Scatter Plot: Relationship between X and Y')

plt.legend()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


data = np.random.randn(1000)


plt.hist(data, bins=30, color='skyblue', edgecolor='black')

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram: Distribution of a Numerical Variable')


plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np


categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = [25, 40, 15, 30, 35]


plt.bar(categories, values, color='skyblue', edgecolor='black')


plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Comparison of Categories Based on Values')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(42)
category_data = {

    'Category A': np.random.normal(25, 5, 100),
    'Category B': np.random.normal(40, 7, 100),
    'Category C': np.random.normal(15, 3, 100),
    'Category D': np.random.normal(30, 4, 100),
    'Category E': np.random.normal(35, 6, 100)
}


plt.boxplot(category_data.values(), vert=True, patch_artist=True, labels=category_data.keys(), 
            boxprops=dict(facecolor='lightblue', color='black'), medianprops=dict(color='red'))

plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Box Plot: Distribution of Values by Category')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


np.random.seed(42)  


categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']

data = {
    'Category A': np.random.normal(10, 5, 100),  # Mean=10, Std=5
    'Category B': np.random.normal(20, 7, 100),  # Mean=20, Std=7
    'Category C': np.random.normal(15, 3, 100),  # Mean=15, Std=3
    'Category D': np.random.normal(30, 8, 100),  # Mean=30, Std=8
    'Category E': np.random.normal(25, 6, 100)   # Mean=25, Std=6
}


import pandas as pd
df = pd.DataFrame({
    'Category': np.repeat(categories, 100),
    'Value': np.concatenate([data[cat] for cat in categories])})

plt.figure(figsize=(8, 6))
sns.boxplot(x='Category', y='Value', data=df, palette='Set2')


plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Distribution of Numerical Values Across Different Categories')


plt.show()
```
